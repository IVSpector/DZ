```python
documents = [
{'type': 'passport', 'number': '2207 876234', 'name': 'Василий Гупкин'},
{'type': 'invoice', 'number': '11-2', 'name': 'Геннадий Покемонов'},
{'type': 'insurance', 'number': '10006', 'name': 'Аристарх Павлов'}
]


directories = {
'1': ['2207 876234', '11-2'],
'2': ['10006'],
'3': []
}


def true_number():
    number_doc = input('Введите номер документа: ')
    for element in documents:
        if number_doc in element['number']:
            return number_doc
    return print('Документ не найден в базе.')

def true_polka():
    number_polka = input('Введите номер полки: ')
    if number_polka in directories:
        return print(f'Такая полка уже существует. Текущий перечень полок: {",".join(a)}')
    return number_polka
 
def empty_polka():
    number_polka = input('Введите номер полки: ')
    if directories.get(str(number_polka)) == []:
        return number_polka
    elif directories.get(str(number_polka)) is None:
        return print ('Такой полки не существует.')
    else:
        return print(f'На полке есть документа, удалите их перед удалением полки. Текущий перечень полок: {",".join(a)}')



def p():
    
    number_doc = true_number()
    if number_doc is None:
        return
    
    for element in documents:
        if element['number'] == number_doc:
            name_owner = element['name'] 
            
    return print(f'Владелец документа: {name_owner}')


def s():
    
    number_doc = true_number()
    if number_doc is None:
        return
    
    for element in directories:
        if number_doc in directories[element]:
            return print (f'Документ хранится на полке: {element}')

        
def l():
    
    for element in documents:
        for number in directories:
            if element["number"] in directories[number]:
                print (f'№ {element["number"]}, тип: {element["type"]}, владелец: {element["name"]}, полка хранения: {number}')
    
def as_func():
    
    number_polka = true_polka()
    if number_polka is None:
        return
    directories[number_polka] = ""
    a = list(directories.keys())
    return print (f'Полка добавлена. Текущий перечень полок: {",".join(a)}')

def ds():
    number_polka = empty_polka()
    if number_polka is None:
        return
    del directories[number_polka]
    a = list(directories.keys())
    return print (f'Полка удалена. Текущий перечень полок: {",".join(a)}')


data = {
    'p': p ,
    's': s ,
    'l': l ,
    'as': as_func ,
    'ds': ds
}




while True:
    command = input('Введите команду (q - выход): ')
    
    if command == 'q':
        break
        
    func = data.get(command)
    if func:
        func()
    else:
        print('Нет такой команды')
```
